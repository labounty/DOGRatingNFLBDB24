```python
def check_indices(df):
    # Get the indices of the rows where 'event' is 'tackle' and 'nflId' is not equal to 'ballCarrierId'
    tackle_indices = df[(df['event'] == 'first_contact') & (df['nflId'] != df['ballCarrierId'])].index

    # Number of frames before 'tackle' event
    X = 10

    # Initialize count
    count = 0

    # List to store the indices where 'game_play_Id' and 'nflId' are different
    different_indices = []

    for idx in tackle_indices:
        # Check if the index X indices before exists in the DataFrame
        if idx - X in df.index:
            # Check if 'game_play_Id' and 'nflId' are the same
            if df.loc[idx, 'game_play_Id'] != df.loc[idx - X, 'game_play_Id'] or df.loc[idx, 'nflId'] != df.loc[idx - X, 'nflId']:
                count += 1
                different_indices.append(idx)

    # Print the indices where 'game_play_Id' and 'nflId' are different
    print('Indices where game_play_Id and nflId are different:', different_indices)

    # Print the count
    print('Count:', count)

    return different_indices
```

### **2.6: Generating Average Magnitude**

```python
def compute_speed_features_original(df1, df2):
    """
    Function to compute averages based on certain conditions and add them as new columns to df1.
    Different conditions are applied based on the value of df1['event_type'].

    Args:
    df1 (pd.DataFrame): Dataframe to iterate through.
    df2 (pd.DataFrame): Dataframe to find the index and calculate averages.

    Returns:
    pd.DataFrame: Updated df1 with new columns.
    """
    # Initialize the new columns in df1
    df1['s_avg_05'] = None
    df1['s_avg_1'] = None
    df1['bc_s_avg_05'] = None
    df1['bc_s_avg_1'] = None

    # Iterate through each row in df1
    for index, row in df1.iterrows():
        # Determine the condition based on df1['event_type']
        if row['event_type'] == 1:
            condition = (df2['event'] == 'tackle') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        elif row['event_type'] == 2:
            condition = (df2['new_event'] == 'missed tackle') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        elif row['event_type'] == 3:
            condition = (df2['event'] == 'fumble') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        else:
            continue  # Skip the row if the event_type is neither 1 nor 2

        matching_indexes = df2.index[condition]

        for i in matching_indexes:
            # Ensure index range does not go below 0
            start_index_05 = max(i - 5, 0)
            start_index_1 = max(i - 10, 0)

            # Calculate averages and assign to df1
            df1.at[index, 's_avg_05'] = df2.loc[start_index_05:i, 's'].mean()
            df1.at[index, 's_avg_1'] = df2.loc[start_index_1:start_index_05, 's'].mean()
            df1.at[index, 'bc_s_avg_05'] = df2.loc[start_index_05:i, 'bc_s'].mean()
            df1.at[index, 'bc_s_avg_1'] = df2.loc[start_index_1:start_index_05, 'bc_s'].mean()

    return df1

# Example usage:
# updated_df1 = compute_averages_modified(df1, df2)
```

```python
def compute_speed_features(df1, df2):
    """
    Function to compute averages based on certain conditions and add them as new columns to df1.
    Different conditions are applied based on the value of df1['event_type'].

    Args:
    df1 (pd.DataFrame): Dataframe to iterate through.
    df2 (pd.DataFrame): Dataframe to find the index and calculate averages.

    Returns:
    pd.DataFrame: Updated df1 with new columns.
    """
    # Initialize the new columns in df1
    df1['s_avg_05'] = None
    df1['s_avg_1'] = None
    df1['bc_s_avg_05'] = None
    df1['bc_s_avg_1'] = None

    # Iterate through each row in df1
    for index, row in df1.iterrows():
        # Print the index if it is a multiple of 1000
        if index % 1000 == 0:
            print(f"Processing index: {index}")
            
        # Determine the condition based on df1['event_type']
        if row['event_type'] == 1:
            condition = (df2['event'] == 'tackle') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        elif row['event_type'] == 2:
            condition = (df2['new_event'] == 'missed tackle') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        elif row['event_type'] == 3:
            condition = (df2['event'] == 'fumble') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        elif row['event_type'] == 4:
            condition = (df2['event'] == 'first_contact') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        else:
            continue  # Skip the row if the event_type is neither 1,2,3, nor 4

        matching_indexes = df2.index[condition]

        #Checking if there are no matching indexes
        if len(matching_indexes) == 0:
            print(f"Index {index} has no matching index")
            continue
        #Checking if there is one matching index (What we want)
        elif len(matching_indexes) == 1:
            i = matching_indexes[0]
        #Checking to make sure there is only one macthing index
        elif len(matching_indexes) > 1:
            print(f"Index {index} has more than one matching index")
    
        # Calculate averages and assign to df1
        df1.at[index, 's_avg_05'] = df2.loc[i-5:i-1, 's'].mean()
        df1.at[index, 's_avg_1'] = df2.loc[i-10:i-6, 's'].mean()
        df1.at[index, 'bc_s_avg_05'] = df2.loc[i-5:i-1, 'bc_s'].mean()
        df1.at[index, 'bc_s_avg_1'] = df2.loc[i-10:i-6, 'bc_s'].mean()

    return df1

# Example usage:
# updated_df1 = compute_averages_modified(df1, df2)
```

### **2.7: Generating Cosine and Alignment Features**

```python
def compute_positional_features(df1, df2):
    """
    Function to compute averages based on certain conditions and add them as new columns to df1.
    Different conditions are applied based on the value of df1['event_type'].

    Args:
    df1 (pd.DataFrame): Dataframe to iterate through.
    df2 (pd.DataFrame): Dataframe to find the index and calculate averages.

    Returns:
    pd.DataFrame: Updated df1 with new columns.
    """
    # Initialize the new columns in df1
    df1['cos_avg_05'] = None
    df1['cos_avg_1'] = None
    df1['do_avg_05'] = None
    df1['do_avg_1'] = None
    df1['bc_do_avg_05'] = None
    df1['bc_do_avg_1'] = None

    # Iterate through each row in df1
    for index, row in df1.iterrows():
        # Print the index if it is a multiple of 1000
        if index % 1000 == 0:
            print(f"Processing index: {index}")

        # Determine the condition based on df1['event_type']
        if row['event_type'] == 1:
            condition = (df2['event'] == 'tackle') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        elif row['event_type'] == 2:
            condition = (df2['new_event'] == 'missed tackle') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        elif row['event_type'] == 3:
            condition = (df2['event'] == 'fumble') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        elif row['event_type'] == 4:
            condition = (df2['event'] == 'first_contact') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        else:
            continue  # Skip the row if the event_type is neither 1,2, nor 3

        matching_indexes = df2.index[condition]

        # Checking if there are no matching indexes
        if len(matching_indexes) == 0:
            print(f"Index {index} has no matching index")
            continue
        # Checking if there is one matching index (What we want)
        elif len(matching_indexes) == 1:
            i = matching_indexes[0]
        # Checking to make sure there is only one matching index
        elif len(matching_indexes) > 1:
            print(f"Index {index} has more than one matching index")

        # Calculate averages and assign to df1
        df1.at[index, 'cos_avg_05'] = df2.loc[i-5:i-1, 'cosine'].mean()
        df1.at[index, 'cos_avg_1'] = df2.loc[i-10:i-6, 'cosine'].mean()
        df1.at[index, 'do_avg_05'] = df2.loc[i-5:i-1, 'do_align'].mean()
        df1.at[index, 'do_avg_1'] = df2.loc[i-10:i-6, 'do_align'].mean()
        df1.at[index, 'bc_do_avg_05'] = df2.loc[i-5:i-1, 'bc_do_align'].mean()
        df1.at[index, 'bc_do_avg_1'] = df2.loc[i-10:i-6, 'bc_do_align'].mean()

    return df1
```
