```python
def check_indices(df):
    # Get the indices of the rows where 'event' is 'tackle' and 'nflId' is not equal to 'ballCarrierId'
    tackle_indices = df[(df['event'] == 'tackle') & (df['nflId'] != df['ballCarrierId'])].index

    # Number of frames before 'tackle' event
    X = 10

    # Initialize count
    count = 0

    # List to store the indices where 'game_play_Id' and 'nflId' are different
    different_indices = []

    for idx in tackle_indices:
        # Check if the index X indices before exists in the DataFrame
        if idx - X in df.index:
            # Check if 'game_play_Id' and 'nflId' are the same
            if df.loc[idx, 'game_play_Id'] != df.loc[idx - X, 'game_play_Id'] or df.loc[idx, 'nflId'] != df.loc[idx - X, 'nflId']:
                count += 1
                different_indices.append(idx)

    # Print the indices where 'game_play_Id' and 'nflId' are different
    print('Indices where game_play_Id and nflId are different:', different_indices)

    # Print the count
    print('Count:', count)

    return different_indices
```

### **2.6: Generating Average Magnitude**

```python
def compute_speed_features(df1, df2):
    """
    Function to compute averages based on certain conditions and add them as new columns to df1.
    Different conditions are applied based on the value of df1['event_type'].

    Args:
    df1 (pd.DataFrame): Dataframe to iterate through.
    df2 (pd.DataFrame): Dataframe to find the index and calculate averages.

    Returns:
    pd.DataFrame: Updated df1 with new columns.
    """
    # Initialize the new columns in df1
    df1['s_avg_05'] = None
    df1['s_avg_1'] = None
    df1['bc_s_avg_05'] = None
    df1['bc_s_avg_1'] = None

    # Iterate through each row in df1
    for index, row in df1.iterrows():
        # Determine the condition based on df1['event_type']
        if row['event_type'] == 1:
            condition = (df2['event'] == 'tackle') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        elif row['event_type'] == 2:
            condition = (df2['new_event'] == 'missed tackle') & \
                        (df2['game_play_Id'] == row['game_play_Id']) & \
                        (df2['nflId'] == row['nflId'])
        else:
            continue  # Skip the row if the event_type is neither 1 nor 2

        matching_indexes = df2.index[condition]

        for i in matching_indexes:
            # Ensure index range does not go below 0
            start_index_05 = max(i - 5, 0)
            start_index_1 = max(i - 10, 0)

            # Calculate averages and assign to df1
            df1.at[index, 's_avg_05'] = df2.loc[start_index_05:i, 's'].mean()
            df1.at[index, 's_avg_1'] = df2.loc[start_index_1:start_index_05, 's'].mean()
            df1.at[index, 'bc_s_avg_05'] = df2.loc[start_index_05:i, 'bc_s'].mean()
            df1.at[index, 'bc_s_avg_1'] = df2.loc[start_index_1:start_index_05, 'bc_s'].mean()

    return df1

# Example usage:
# updated_df1 = compute_averages_modified(df1, df2)
```

```python
###Checking to see if blank values are due to player being pushed out of bounds
def count_out_of_bounds(tackles_1v1_1_test, tracking_clean_1):
    # Filter rows where s_avg_05 is None
    filtered_tackles = tackles_1v1_1_test[tackles_1v1_1_test['s_avg_05'].isnull()]

    count = 0
    # Iterate over the filtered rows
    for _, row in filtered_tackles.iterrows():
        # Find matching rows in tracking_clean_1
        matching_rows = tracking_clean_1[(tracking_clean_1['game_play_Id'] == row['game_play_Id']) & 
                                         (tracking_clean_1['nflId'] == row['nflId'])]
        # Count how many times 'event' is 'out_of_bounds'
        count += (matching_rows['event'] == 'out_of_bounds').sum()
    
    return count
```

```python
def calculate_speed_features(df):
    # Create an empty DataFrame to store the speed features
    speed_features = pd.DataFrame()

    # Iterate over each observation in df
    for index, row in df.iterrows():
        if row['event_type'] == 1:
            # Find the index 'i' in df
            i = df[(df['event'] == 'tackle') & 
                   (df['game_play_Id'] == row['game_play_Id']) & 
                   (df['nflId'] == row['nflId'])].index[0]
            
            # Calculate the averages
            s_05_avg, s_1_avg, bc_s_05_avg, bc_s_1_avg = calculate_averages(df, i)
            
        elif row['event_type'] == 2:
            # Find the index 'i' in df
            i = df[(df['missed_event'] == 'missed tackle') & 
                   (df['game_play_Id'] == row['game_play_Id']) & 
                   (df['nflId'] == row['nflId'])].index[0]
            
            # Calculate the averages
            s_05_avg, s_1_avg, bc_s_05_avg, bc_s_1_avg = calculate_averages(df, i)
        
        # Append the speed features to the speed_features DataFrame
        speed_features = speed_features.append({'s_05': s_05_avg, 's_1': s_1_avg, 'bc_s_05': bc_s_05_avg, 'bc_s_1': bc_s_1_avg}, ignore_index=True)

    return speed_features

def calculate_averages(df, i):
    # Calculate the average of df[s] from i-5:i
    s_05_avg = df.loc[i-5:i, 's'].mean()
    
    # Calculate the average of df[s] from i-10:i-5
    s_1_avg = df.loc[i-10:i-5, 's'].mean()
    
    # Calculate the average of df[bc_s] from i-5:i
    bc_s_05_avg = df.loc[i-5:i, 'bc_s'].mean()
    
    # Calculate the average of df[bc_s] from i-10:i-5
    bc_s_1_avg = df.loc[i-10:i-5, 'bc_s'].mean()

    return s_05_avg, s_1_avg, bc_s_05_avg, bc_s_1_avg
```
